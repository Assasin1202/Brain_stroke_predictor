```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import (
    RandomForestClassifier, GradientBoostingClassifier, 
    AdaBoostClassifier, ExtraTreesClassifier
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, 
    roc_curve, accuracy_score, precision_score, recall_score, f1_score
)

from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
import shap


RANDOM_STATE = 42

# -------------------------------
# 1. Data Loading and Exploration
# -------------------------------
data_path = "resampled-stroke-data.csv"
df = pd.read_csv(data_path)

if df['bmi'].isnull().sum() > 0:
    df['bmi'].fillna(df['bmi'].median(), inplace=True)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x="age", hue="stroke", multiple="stack", bins=30)
plt.title("Age Distribution by Stroke Status")

plt.subplot(1, 2, 2)
sns.histplot(data=df, x="bmi", hue="stroke", multiple="stack", bins=30)
plt.title("BMI Distribution by Stroke Status")
plt.tight_layout()
plt.show()

# -------------------------------
# 2. Data Preprocessing
# -------------------------------
X = df.drop(columns=["stroke"])
y = df["stroke"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)
print("\nTraining set shape:", X_train.shape)
print("Test set shape:", X_test.shape)

continuous_features = ["age", "avg_glucose_level", "bmi"]
scaler = StandardScaler()
X_train[continuous_features] = scaler.fit_transform(X_train[continuous_features])
X_test[continuous_features] = scaler.transform(X_test[continuous_features])

# -------------------------------
# 3. Model Training and Evaluation
# -------------------------------

models = {
    "Logistic Regression": LogisticRegression(random_state=RANDOM_STATE, max_iter=1000),
    "Gaussian NB": GaussianNB(),
    "Random Forest": RandomForestClassifier(random_state=RANDOM_STATE),
    "Decision Tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "SVM": SVC(probability=True, random_state=RANDOM_STATE),
    "Gradient Boosting": GradientBoostingClassifier(random_state=RANDOM_STATE),
    "AdaBoost": AdaBoostClassifier(random_state=RANDOM_STATE),
    "Extra Trees": ExtraTreesClassifier(random_state=RANDOM_STATE),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "MLP Classifier": MLPClassifier(hidden_layer_sizes=(16,), max_iter=2000, random_state=RANDOM_STATE),
    "LightGBM": LGBMClassifier(n_estimators=50, max_depth=3, random_state=RANDOM_STATE),
    "XGBoost": XGBClassifier(n_estimators=50, max_depth=3, use_label_encoder=False, eval_metric="logloss", random_state=RANDOM_STATE)
}

results = []

def plot_confusion_matrix(cm, model_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu")
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

def plot_classification_report_heatmap(report_dict, model_name):
    df_report = pd.DataFrame(report_dict).transpose().iloc[:-1, :-1]
    plt.figure(figsize=(8, 4))
    sns.heatmap(df_report, annot=True, cmap="YlOrRd")
    plt.title(f"{model_name} - Classification Report")
    plt.show()



for model_name, model in models.items(): # Let us evaluate all models
    print("\n\n\n----------------------------------------------------")
    print(f"Training: {model_name}")

    model.fit(X_train, y_train) # Training the model. 

    y_pred = model.predict(X_test)
    
    report = classification_report(y_test, y_pred, output_dict=True)
    cm = confusion_matrix(y_test, y_pred)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    roc_auc = None
    fpr = None
    tpr = None
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_prob)
        fpr, tpr, _ = roc_curve(y_test, y_prob)
    
    print(f"\nClassification Report for {model_name}:")
    print(classification_report(y_test, y_pred))
    plot_confusion_matrix(cm, model_name)
    plot_classification_report_heatmap(report, model_name)
    
    if roc_auc is not None:
        plt.figure(figsize=(6, 5))
        plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.4f})")
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"{model_name} - ROC Curve")
        plt.legend()
        plt.show()
    else:
        print(f"{model_name} does not support probability estimates for ROC AUC.")
    

    # ---------------------------
    # 4. Model Size Calculation
    # ---------------------------

    model_bytes = pickle.dumps(model) 
    size_kb = len(model_bytes) / 1024
    print(f"Model size for {model_name}: {size_kb:.2f} KB")
    print(f"The model {model_name} has been processed.")
    print("Model metrices are saved in the results list.")
    print("\n--")
    
    
    # -----------------------------
    # 5. Saving the Results
    # -----------------------------
    
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC AUC": roc_auc if roc_auc is not None else np.nan,
        "Model Size (KB)": size_kb
    })
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

results_df = pd.DataFrame(results)

models = results_df["Model"]
accuracy = results_df["Accuracy"]
f1_scores = results_df["F1 Score"]
model_sizes = results_df["Model Size (KB)"]

sns.set_theme(style="whitegrid")

custom_colors = [
    "#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B2",
    "#937860", "#DA8BC3", "#8C8C8C", "#CCB974", "#64B5CD",
    "#FFB5B8", "#9EC79D", "#B0A8B9", "#F2C078", "#E07A5F",
    "#B56576", "#6D597A", "#355070", "#F8B195", "#C06C84"
]

def plot_bar_chart(metric, title, ylabel):
    fig, ax = plt.subplots(figsize=(12, 8))
    
    colors = custom_colors[:len(models)]
    
    bars = ax.bar(models, metric, color=colors, edgecolor='black', linewidth=1.2)
    
    ax.set_title(title, fontsize=18, fontweight='bold', pad=15)
    ax.set_xlabel("Models", fontsize=14, labelpad=10)
    ax.set_ylabel(ylabel, fontsize=14, labelpad=10)
    
    ax.tick_params(axis='x', rotation=45, labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    ax.yaxis.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

plot_bar_chart(accuracy,    "Model Accuracy Comparison",     "Accuracy")
plot_bar_chart(f1_scores,   "Model F1 Score Comparison",     "F1 Score")
plot_bar_chart(model_sizes, "Model Size Comparison (in KB)", "Model Size (KB)")
```

```python
import os
import numpy as np
import onnxruntime as ort
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType
from onnxruntime.quantization import quantize_dynamic, QuantType
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings("ignore")

def plot_classification_report_heatmap(report_dict, model_name):
    df_report = pd.DataFrame(report_dict).transpose().iloc[:-1, :-1]
    plt.figure(figsize=(8, 4))
    sns.heatmap(df_report, annot=True, cmap="YlOrRd")
    plt.title(f"{model_name} - Classification Report")
    plt.show()


# -----------------
# 6. Quantization of the model to reduce size
# -----------------

big_mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=2000, random_state=RANDOM_STATE)
big_mlp.fit(X_train, y_train)

y_pred_big = big_mlp.predict(X_test)
print("Classification Report for Big MLP (scikit-learn):")



report_big = classification_report(y_test, y_pred_big, output_dict=True)
plot_classification_report_heatmap(report_big, "Big MLP (scikit-learn)")

initial_type = [('float_input', FloatTensorType([None, X_train.shape[1]]))]
onnx_big_model = convert_sklearn(big_mlp, initial_types=initial_type)
onnx_big_model_path = "big_mlp_model.onnx"
with open(onnx_big_model_path, "wb") as f:
    f.write(onnx_big_model.SerializeToString())
print("Big MLP model converted to ONNX format and saved as:", onnx_big_model_path)


quantized_big_model_path = "big_mlp_model_quant.onnx"
quantize_dynamic(onnx_big_model_path, quantized_big_model_path, weight_type=QuantType.QInt8)
print("Quantized Big MLP model saved as:", quantized_big_model_path)


original_size = os.path.getsize(onnx_big_model_path) / 1024
quantized_size = os.path.getsize(quantized_big_model_path) / 1024
print("\n\nOriginal Big MLP ONNX model size: {:.2f} KB".format(original_size))
print("Quantized Big MLP ONNX model size: {:.2f} KB".format(quantized_size))

```

```python
sess = ort.InferenceSession(quantized_big_model_path)
input_name = sess.get_inputs()[0].name

X_test_np = X_test.astype(np.float32).values

y_pred_quant_big = sess.run(None, {input_name: X_test_np})[0]

print("Classification Report for Quantized Big MLP (ONNX Runtime):")
report_quant_big = classification_report(y_test, y_pred_quant_big, output_dict=True)
plot_classification_report_heatmap(report_quant_big, "Quantized Big MLP (ONNX Runtime)")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lime.lime_tabular
import shap

# -------------------------------
# LIME and SHAP Explainers
# -------------------------------

lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_train.values,
    feature_names=X_train.columns.tolist(),
    class_names=["No Stroke", "Stroke"],
    mode='classification',
    discretize_continuous=False
)

background_sample = X_train.sample(50, random_state=42)
shap_explainer = shap.KernelExplainer(
    model=big_mlp.predict_proba,
    data=background_sample
)

# -------------------------------
# Function to Compute, Print, and Plot Explanations for a Given Instance
# -------------------------------
def explain_sample(instance_index):
    """
    For the given test instance index, this function:
      - Prints the original unscaled sample (with column names).
      - Computes and prints the LIME explanation (for the 'Stroke' class).
      - Plots the LIME explanation.
      - Computes and prints the SHAP explanation (for the 'Stroke' class).
      - Plots the SHAP force plot.
    """
    
    print(f"\n=== Test Sample (Index {instance_index}) ===")
    print(df.loc[instance_index])
    

    instance_array = X_test.loc[instance_index].values
    
    # ---- LIME Explanation ----
    lime_exp = lime_explainer.explain_instance(
        data_row=instance_array,
        predict_fn=big_mlp.predict_proba,
        num_features=8,
        labels=(0, 1)
    )
    print(f"\n--- LIME Explanation for Instance Index: {instance_index} (Class: 'Stroke') ---")
    for feature, weight in lime_exp.as_list(label=1):
        print(f"Feature: {feature:<40} Weight: {weight:.4f}")
    
    # Plot LIME explanation
    fig_lime = lime_exp.as_pyplot_figure()
    plt.title(f"LIME Explanation (Class='Stroke') - Instance #{instance_index}")
    plt.show()
    
    # ---- SHAP Explanation ----

    shap_values_local = shap_explainer.shap_values(X_test.loc[[instance_index]])
    local_contrib_class1 = shap_values_local[1][0]  # For class 'Stroke'
    print(f"\n--- SHAP Explanation for Instance Index: {instance_index} (Class: 'Stroke') ---")

    sorted_idx = np.argsort(np.abs(local_contrib_class1))[::-1]
    for idx in sorted_idx:
        feat_name = X_test.columns[idx]
        feat_value = X_test.loc[instance_index, feat_name]
        shap_val = local_contrib_class1[idx]
        print(f"Feature: {feat_name:<20} Value: {feat_value:.3f}  SHAP: {shap_val:.4f}")
    
# -------------------------------
# Compute Explanations for 5 Random Samples
# -------------------------------

random_sample_indices = np.random.choice(X_test.index, size=5, replace=False)
print("=== Explanations for 5 Random Test Samples ===")
for idx in random_sample_indices:
    print("\n--------------------------------------------------")
    explain_sample(idx)
    print("--------------------------------------------------\n")
    
# -------------------------------
# Overall SHAP Values for 50 Instances (Global Explanation)
# -------------------------------

sample_size = 50
shap_values_global = shap_explainer.shap_values(X_test.iloc[:sample_size])

print("=== SHAP Summary Plot (Bar Plot) for 50 Test Instances (Class: 'Stroke') ===")
shap.summary_plot(
    shap_values_global[1],
    X_test.iloc[:sample_size],
    feature_names=X_test.columns,
    plot_type='bar'
)
```
