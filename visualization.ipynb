```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "healthcare-dataset-stroke-data.csv"  # Update this path if necessary
df = pd.read_csv(file_path)

# Check the class distribution
stroke_counts = df['stroke'].value_counts()

# Display class distribution numerically
stroke_counts_df = pd.DataFrame({'Stroke': stroke_counts.index, 'Count': stroke_counts.values})
print("Class Distribution of Stroke:")
print(stroke_counts_df)

# Plot class distribution
plt.figure(figsize=(6, 4))
sns.barplot(x=stroke_counts.index, y=stroke_counts.values, palette="viridis")
plt.xticks(ticks=[0, 1], labels=["No Stroke (0)", "Stroke (1)"])
plt.ylabel("Count")
plt.xlabel("Stroke Occurrence")
plt.title("Class Imbalance in Stroke Prediction Dataset")
plt.show()
```

### Age distribution by Stroke. 

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(df[df['stroke'] == 0]['age'], label="No Stroke", color='blue', shade=True)
sns.kdeplot(df[df['stroke'] == 1]['age'], label="Stroke", color='red', shade=True)
plt.title("Age Distribution by Stroke Status")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
plt.show()
```

### BMI Distribution

```python
plt.figure(figsize=(10, 6))
sns.distplot(df[df['stroke'] == 0]['bmi'], label="No Stroke", color='green', kde=True)
sns.distplot(df[df['stroke'] == 1]['bmi'], label="Stroke", color='red', kde=True)
plt.title("BMI Distribution by Stroke Status")
plt.xlabel("BMI")
plt.ylabel("Density")
plt.legend()
plt.show()
```
